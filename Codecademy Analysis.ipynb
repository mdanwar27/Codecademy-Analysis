**Filled null values in the "Certificate" column**

```python
df["Certificate"]=df["Certificate"].fillna("No Certificate")
```

**Splitting each string into words, filtering out any occurrence of the word "With", and then joining the remaining words back into a single string**

```python
df["Certificate"] = df["Certificate"].apply(lambda x: " ".join([word for word in x.split() if word != "With"]))
```

**Splitting each string into words, filtering out any occurrence of the word "Friendly", and then joining the remaining words back into a single string**

```python
df["Level"] = df["Level"].apply(lambda x: " ".join([word for word in x.split() if word != "Friendly"]))
```

**Extracted only numerical values in the "Enrolled" column using regex**

```python
df["Enrolled"]=df["Enrolled"].str.extract(r"(\d{1,3}(?:,\d{3})*)")
```

**Replaced the "," in the "Enrolled" column with empty space**

```python
df["Enrolled"]=df["Enrolled"].str.replace(",","")
```

**The reason to fill null values with 0 in enrolled column is in the next 4 cells**

```python
df[df["Enrolled"].isna()]
```

```python
df["Path"].value_counts()
```

```python
df[(df["Enrolled"].notnull()) & (df["Path"]=="Free course")]
```

**Filled null values in the "Enrolled" column and converted the datatype into int**

```python
df["Enrolled"]=df["Enrolled"].fillna(0).astype(int)
```

**Extracted only numerical values in the "Ratings" column using regex**

```python
df["Ratings"]=df["Ratings"].str.extract(r"(.\d+)")
```

**Replacing comma with empty space and filling null values with 0 and converting the datatype to int**

```python
df["Ratings"]=df["Ratings"].str.replace(",","")
df["Ratings"]=df["Ratings"].fillna(0).astype(int)
```

**Filling null values in the Review column with 0**

```python
df["Review"]=df["Review"].fillna(0)
```

**Categorizing rows in the DataFrame based on their index values using predefined ranges, and then creating a new column (Category) that indicates the category each row belongs to**  
**Example: Index 50 would fall into the range 0 to 173, hence mapped to "Web Development"**

```python
category_mapping = {
    'Artificial Intelligence': range(0, 98),
    'Bash/Shell': range(98, 111),
    'C': range(111, 118),
    'C#': range(118, 127),
    'C++': range(127, 135),
    'Cloud Computing': range(135, 222),
    'Code Foundations': range(222, 250),
    'Computer Science': range(250, 379),
    'Cyber Security': range(379, 450),
    'Data Analytics': range(450, 495),
    'Data Engineering': range(495, 500),
    'Data Science': range(500, 616),
    'Data Visualization': range(616, 637),
    'Developer Tools': range(637, 642),
    'DevOps': range(642, 654),
    'Game Development': range(654, 661),
    'Go': range(661, 668),
    'HTML & CSS': range(668, 699),
    'Interview Prep': range(699, 711),
    'IT': range(711, 723),
    'Java': range(723, 751),
    'JavaScript': range(751, 846),
    'Kotlin': range(846, 850),
    'Machine Learning': range(850, 894),
    'Math': range(894, 914),
    'Mobile Development': range(914, 942),
    'Open Source': range(942, 954),
    'PHP': range(954, 962),
    'Professional Skills': range(962, 1032),
    'Python': range(1032, 1151),
    'R': range(1151, 1165),
    'Ruby': range(1165, 1168),
    'SQL': range(1168, 1186),
    'Swift': range(1186, 1204),
    'Web Design': range(1204, 1238),
    'Web Development': range(1238, 1412)
    
}

def map_category(index):
    for category, index_range in category_mapping.items():
        if index in index_range:
            return category
    return 'Other'
df["Category"] = df.index.map(map_category)

df
```

**As my target variable is Enrolled column, Identifying outliers in the column**

```python
Q1=df["Enrolled"].quantile(0.25)
Q3=df["Enrolled"].quantile(0.75)
IQR=Q3-Q1
print("Q1:", Q1)
print("Q3:", Q3)
upper_bound=Q3 + 1.5*IQR
lower_bound=Q1 - 1.5*IQR
print("Upper bound:",upper_bound)
print("Lower bound:",lower_bound)
```

```python
outliers_df = df[(df["Enrolled"] < lower_bound) | (df["Enrolled"] > upper_bound)]
non_outliers_df = df[(df["Enrolled"] >= lower_bound) & (df["Enrolled"] <= upper_bound)]
```

## Unvariate Analysis for Non outlier dataframe

```python
from wordcloud import WordCloud
# Word Cloud: Visualize the most common words in Path and Title.

path_counts = non_outliers_df['Path'].value_counts()
print("Types of Paths:")
print()
print(path_counts)

# Word cloud for 'Path'
wordcloud_path = WordCloud(width=800, height=400, background_color='white').generate(' '.join(non_outliers_df['Path']))
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_path, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Paths')
plt.show()
```
